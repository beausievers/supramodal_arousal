# Emotional free-drawing survey analyzed with Harris corner detection and Bayesian logistic regression

```python
%matplotlib inline

import glob

import numpy as np

import harris_corners as hc
import bayesian_logistic as bl
```

## Load the data

```python
base_path = '../data/study_2/drawing_example/'
```

```python
angry_paths = glob.glob(base_path + 'angry/*.tif')
angry_corner_counts = []
for path in angry_paths:
    angry_corner_counts.append(hc.count_corners(path))
```

```python
sad_paths = glob.glob(base_path + 'sad/*.tif')
sad_corner_counts = []
for path in sad_paths:
    sad_corner_counts.append(hc.count_corners(path))
```

```python
corner_counts = angry_corner_counts + sad_corner_counts
angry = [True] * len(angry_corner_counts) + [False] * len(sad_corner_counts)
```

## Run Bayesian logistic regression

```python
reload(bl)
```

```python
X = np.array(corner_counts)
y = np.array(angry)
```

```python
my_bl_cv = bl.BayesianLogisticClassifierCV(X, y, n_folds=5, n_iter=5000, n_chains=4)
```

```python
my_bl_cv.fit()
```

```python
my_bl_cv.plot_all(
    true_cat="Angry", 
    false_cat="Sad",
    title="Survey drawings with prompt image",
    reg_x_label="Corner count",
    reg_y_label="probability(Angry)",
    small=False
)
```
